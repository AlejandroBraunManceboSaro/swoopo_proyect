# Categorización los prodructos a través de Amazon

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import re
```

```python
outcomes_df=pd.read_csv("C:/Users/Alex/OneDrive/Desktop/python_proyects/outcomes.tsv", sep="\t")
```

```python
productos=outcomes_df.item.unique()
```

```python

dict_categorias={}
for product in productos:
    path= "C:/Users/Alex/OneDrive/Desktop/python_proyects/Chrome_driver/chromedriver_win32/chromedriver"
    service=Service(executable_path=path)
    driver = webdriver.Chrome(service=service)
    product_key=product
    product=product.replace("-","+")
    if re.search("[+]$",product):
        product=product.rstrip(product[-1])
    ##### Opciones de split
    product_splited=product.split("+")
    product_1out=" ".join(product_splited[:-1])
    product_2out=" ".join(product_splited[:-2])
    product_3out=" ".join(product_splited[:-3])
    product_4out=" ".join(product_splited[:-4])
    product_5out=" ".join(product_splited[:-5])
    product_6out=" ".join(product_splited[:-6])
    product_1word=" ".join(product_splited[:-1000])
    
    try:
        web=f'https://www.amazon.es/s?k={product}&__mk_es_US'
        driver.get(web)
        link=driver.find_element(by="xpath",value='//*[@id="sp-cc-accept"]')
        link.click()
        element = WebDriverWait(driver,5).until(EC.presence_of_element_located((By.XPATH,'//div[@id="departments"]')))
        element=element.text
        element=element.split("\n")[1]
        dict_categorias[product_key]=element
        driver.quit()
    except:
        try:
            path= "C:/Users/Alex/OneDrive/Desktop/python_proyects/Chrome_driver/chromedriver_win32/chromedriver"
            service=Service(executable_path=path)
            driver = webdriver.Chrome(service=service)
            web=f'https://www.amazon.es/s?k={product_1out}'
            driver.get(web)
            link=driver.find_element(by="xpath",value='//*[@id="sp-cc-accept"]')
            link.click()
            element = WebDriverWait(driver,5).until(EC.presence_of_element_located((By.XPATH,'//div[@id="departments"]')))
            element=element.text
            element=element.split("\n")[1]
            dict_categorias[product_key]=element
            driver.quit()
        except:
            try:
                path= "C:/Users/Alex/OneDrive/Desktop/python_proyects/Chrome_driver/chromedriver_win32/chromedriver"
                service=Service(executable_path=path)
                driver = webdriver.Chrome(service=service)
                web=f'https://www.amazon.es/s?k={product_2out}'
                driver.get(web)
                link=driver.find_element(by="xpath",value='//*[@id="sp-cc-accept"]')
                link.click()
                element = WebDriverWait(driver,5).until(EC.presence_of_element_located((By.XPATH,'//div[@id="departments"]')))
                element=element.text
                element=element.split("\n")[1]
                dict_categorias[product_key]=element
                driver.quit()
            except:
                try:
                    path= "C:/Users/Alex/OneDrive/Desktop/python_proyects/Chrome_driver/chromedriver_win32/chromedriver"
                    service=Service(executable_path=path)
                    driver = webdriver.Chrome(service=service)
                    web=f'https://www.amazon.es/s?k={product_3out}'
                    driver.get(web)
                    link=driver.find_element(by="xpath",value='//*[@id="sp-cc-accept"]')
                    link.click()
                    element = WebDriverWait(driver,5).until(EC.presence_of_element_located((By.XPATH,'//div[@id="departments"]')))
                    element=element.text
                    element=element.split("\n")[1]
                    dict_categorias[product_key]=element
                    driver.quit()
                except:
                    try:
                        path= "C:/Users/Alex/OneDrive/Desktop/python_proyects/Chrome_driver/chromedriver_win32/chromedriver"
                        service=Service(executable_path=path)
                        driver = webdriver.Chrome(service=service)
                        web=f'https://www.amazon.es/s?k={product_4out}'
                        driver.get(web)
                        link=driver.find_element(by="xpath",value='//*[@id="sp-cc-accept"]')
                        link.click()
                        element = WebDriverWait(driver,5).until(EC.presence_of_element_located((By.XPATH,'//div[@id="departments"]')))
                        element=element.text
                        element=element.split("\n")[1]
                        dict_categorias[product_key]=element
                        driver.quit()
                    except:
                        try:
                            path= "C:/Users/Alex/OneDrive/Desktop/python_proyects/Chrome_driver/chromedriver_win32/chromedriver"
                            service=Service(executable_path=path)
                            driver = webdriver.Chrome(service=service)
                            web=f'https://www.amazon.es/s?k={product_5out}'
                            driver.get(web)
                            link=driver.find_element(by="xpath",value='//*[@id="sp-cc-accept"]')
                            link.click()
                            element = WebDriverWait(driver,5).until(EC.presence_of_element_located((By.XPATH,'//div[@id="departments"]')))
                            element=element.text
                            element=element.split("\n")[1]
                            dict_categorias[product_key]=element
                            driver.quit()
                        except:
                            try:
                                path= "C:/Users/Alex/OneDrive/Desktop/python_proyects/Chrome_driver/chromedriver_win32/chromedriver"
                                service=Service(executable_path=path)
                                driver = webdriver.Chrome(service=service)
                                web=f'https://www.amazon.es/s?k={product_6out}'
                                driver.get(web)
                                link=driver.find_element(by="xpath",value='//*[@id="sp-cc-accept"]')
                                link.click()
                                element = WebDriverWait(driver,5).until(EC.presence_of_element_located((By.XPATH,'//div[@id="departments"]')))
                                element=element.text
                                element=element.split("\n")[1]
                                dict_categorias[product_key]=element
                                driver.quit()
                            except:
                                try:
                                    path= "C:/Users/Alex/OneDrive/Desktop/python_proyects/Chrome_driver/chromedriver_win32/chromedriver"
                                    service=Service(executable_path=path)
                                    driver = webdriver.Chrome(service=service)
                                    web=f'https://www.amazon.es/s?k={product_1word}'
                                    driver.get(web)
                                    link=driver.find_element(by="xpath",value='//*[@id="sp-cc-accept"]')
                                    link.click()
                                    element = WebDriverWait(driver,5).until(EC.presence_of_element_located((By.XPATH,'//div[@id="departments"]')))
                                    element=element.text
                                    element=element.split("\n")[1]
                                    dict_categorias[product_key]=element
                                    driver.quit()
                                except:
                                    dict_categorias[product_key]="fallo dell sitema"
                                    driver.quit()
```
